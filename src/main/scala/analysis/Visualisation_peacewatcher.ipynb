```
%scala
import org.apache.spark.sql.SparkSession


val blobName = "wasbs://peacestate@peacestate.blob.core.windows.net/topics/reports/"

val spark = SparkSession.builder()
  .appName("Lecture des données depuis Azure Blob Storage")
  .getOrCreate()

spark.conf.set("fs.azure.account.auth.type.peacestate.blob.core.windows.net", "SharedKey")
spark.conf.set("fs.azure.account.key.peacestate.blob.core.windows.net", storageAccountKey)

val df = spark.read.format("json").load(blobName)
val dfReports = df.drop("namesCitizen")

display(dfReports)

```

```
%scala

import org.apache.spark.sql.functions._

//nb d'alertes moyen par jour de semaine
val dfDayOfWeek = dfAlerts.withColumn("dayOfWeek", date_format(to_date(col("currentDate"), "dd-MM-yyyy"), "EEEE"))

val avgAlerts = dfDayOfWeek.groupBy("dayOfWeek").agg(avg("peaceScore").as("avgAlerts"))
display(avgAlerts)
```

```
%scala
import org.apache.spark.sql.functions._

// nombre d'alertes par mois 
val dfWithMonth = dfAlerts.withColumn("month", date_format(to_date(col("currentDate"), "dd-MM-yyyy"), "MMMM"))

val alertsByMonth = dfWithMonth.groupBy("month").count()
display(alertsByMonth)

```

```
%scala
// TOP 10 des emotions les plus courantes
val wordCounts = dfReports.groupBy("wordHeard").count().orderBy(desc("count")).limit(10)
display(wordCounts)
```

```
%scala
//citoyens avec le meilleur peace score 
val topCitizens = dfReports.filter(col("peaceScore") === 100).select("nameCitizen", "peaceScore", "latitude", "longitude")
display(topCitizens)
```
